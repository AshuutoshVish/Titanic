```python
#Splitting data
```

```python
x = df.drop(columns = "Performance Index")
y= df['Performance Index']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
len(x_train), len(y_train), len(y_test), len(y_test), len(x), len(y)
```

```python
model_mlr=LinearRegression()
```

```python
model_mlr.fit(x_train, y_train)
```

```python
y_pred_mlr = model_mlr.predict(x_test)
y_pred_mlr
```

```python
mae_mlr = mean_absolute_error(y_test, y_pred_mlr)
mae_mlr
```

```python
mse_mlr= mean_squared_error(y_test, y_pred_mlr)
mse_mlr
```

```python
r2_mlr= r2_score(y_test, y_pred_mlr)
r2_mlr
```

```python
mae_mlr = mean_absolute_error(y_test, y_pred_mlr)
mse_mlr = mean_squared_error(y_test, y_pred_mlr)
r2_mlr = r2_score(y_test, y_pred_mlr)

print(f"Mean Absolute Error: {mae_mlr}")
print(f"Mean Squared Error: {mse_mlr}")
print(f"R-Squared Score: {r2_mlr}")
```

```python
print(f"Intercept: {model_mlr.intercept_}")
print(f"Coefficients: {model_mlr.coef_}")
```

```python
model_mlr.score(x_train, y_train)
```

```python
plt.scatter(y_test, y_pred_mlr)
plt.show()
```
